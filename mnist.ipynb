```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split

digits = load_digits()
X, y = digits.data, digits.target

# Divida os dados em treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
X_train
```

```python
from umap import UMAP

reducer = UMAP(n_components=2)
y_train_umap = reducer.fit_transform(X_train)
y_test_umap = reducer.fit_transform(X_test)
```

```python
import tensorflow as tf
from sklearn.model_selection import cross_val_score

input_size = 64
output_size = 2
learning_rate = 0.001
num_epochs = 50
batch_size = 16

# Define o modelo
model = tf.keras.Sequential([
    tf.keras.layers.Dense(256, activation='relu', input_shape=(input_size,)),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(output_size, activation='linear')
])

# Define a função de perda e o otimizador
loss_fn = tf.keras.losses.MeanSquaredError()
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

# Compila o modelo
model.compile(optimizer=optimizer, loss=loss_fn, metrics=['accuracy'])
```

```python
#obs: input_size = 64
# Treina o modelo
model.fit(X_train, y_train_umap, epochs=num_epochs, batch_size=batch_size, validation_data=(X_test, y_test_umap))

# Avalia o modelo
loss, accuracy = model.evaluate(X_test, y_test_umap)
print(f'Test accuracy: {accuracy:.4f}')
```

```python
colors = []
for number in y_test:
  colors.append(str(number))
```

```python
def get_coords(X):
  xcoord = []
  ycoord = []
  for value in X:
    xcoord.append(value[0])
    ycoord.append(value[1])
  coords = [xcoord,ycoord]
  return coords
```

```python
MLP_reduced = model.predict(X_test)
coordsMLP = get_coords(MLP_reduced)
```

```python
import plotly.express as px

fig = px.scatter(x=coordsMLP[0], y=coordsMLP[1], color=colors,height=700,width=950, title="MLP")
fig.show()
```
